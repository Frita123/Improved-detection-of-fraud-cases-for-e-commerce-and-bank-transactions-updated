### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
```

### Load Dataset

```python
credit_df = pd.read_csv(
    "../data/raw/creditcard.csv"
)

credit_df.head()
```

### Dataset Overview

```python
credit_df.shape
credit_df.info()
credit_df.describe()
```

### Data Cleaning

##### Missing Values

```python
credit_df.isnull().sum().sum()
```

##### Duplicate Records

```python
credit_df.duplicated().sum()
credit_df.drop_duplicates(inplace=True)
```

```python
credit_df.dtypes
```

### Exploratory Data Analysis (EDA)

### Class Distribution

```python
class_counts = credit_df['Class'].value_counts()
class_counts
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Class', data=credit_df)
plt.title("Class Distribution (0 = Legit, 1 = Fraud)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

### Percentage of Fraud

```python
fraud_percentage = class_counts[1] / class_counts.sum() * 100
print(f"Fraud transactions: {fraud_percentage:.3f}%")
```

### Transaction Amount Distribution

```python
plt.figure(figsize=(6,4))
sns.histplot(credit_df['Amount'], bins=50, kde=True)
plt.title("Transaction Amount Distribution")
plt.show()
```

### Amount vs Class

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Class', y='Amount', data=credit_df)
plt.title("Transaction Amount by Class")
plt.show()
```

### Time Feature Analysis

```python
plt.figure(figsize=(6,4))
sns.histplot(credit_df['Time'], bins=50)
plt.title("Transaction Time Distribution")
plt.show()
```

### Top Users by Fraud Percentage

```python
credit_df['user_proxy'] = credit_df[['V1', 'V2', 'V3']].apply(lambda x: hash(tuple(x)), axis=1)

# Compute fraud percentage per user
fraud_by_user = credit_df.groupby('user_proxy')['Class'].mean().sort_values(ascending=False)

print("Top 10 users by fraud percentage:")
display(fraud_by_user.head(10))

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=fraud_by_user.head(10).index.astype(str), y=fraud_by_user.head(10).values)
plt.xticks(rotation=45)
plt.title("Top 10 Users by Fraud Percentage")
plt.ylabel("Fraud Percentage")
plt.xlabel("User Proxy")
plt.show()
```

### Correlation Analysis (Sampled)

```python
sample_df = credit_df.sample(50000, random_state=42)

plt.figure(figsize=(12,8))
sns.heatmap(
    sample_df.corr(),
    cmap="coolwarm",
    center=0,
    linewidths=0.1
)
plt.title("Feature Correlation Heatmap (Sampled)")
plt.show()
```

### Save Cleaned Dataset

```python
import os

processed_path = "../data/processed"
os.makedirs(processed_path, exist_ok=True)

credit_df.to_csv(
    os.path.join(processed_path, "creditcard_cleaned.csv"),
    index=False
)

print("Cleaned credit card data saved successfully.")
```
